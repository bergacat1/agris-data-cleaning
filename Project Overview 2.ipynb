```python
import os
os.environ["PYSPARK_PYTHON"] = "python3"
os.environ['PYSPARK_SUBMIT_ARGS'] = ('--packages '
    'graphframes:graphframes:0.3.0-spark2.0-s_2.11'
    ' pyspark-shell')

from pyspark import SparkContext, SparkConf
from pyspark.sql import SQLContext

sc = SparkContext('local[*]')
sqlContext = SQLContext(sc)
```

```python
coauthorsunique = sc.textFile("data/coauthorsunique/Zidorn/part*").map(lambda x: x.split(";"))
coauthorslist = sc.textFile("data/coauthorslist/Zidorn/part*").map(lambda x: x.split(";"))
```

```python
v = sqlContext.createDataFrame(coauthorsunique,['id','times'])
e = sqlContext.createDataFrame(coauthorslist,['src','dst','times'])
```

```python
#v.collect()
```

```python
#e.collect()
```

```python
from graphframes import *
```

```python
g = GraphFrame(v, e)
```

```python
!pip install toyplot
```

```python
import numpy as np
edges_array = np.array(
    g.edges
    .select("src", "dst").rdd
    .map(lambda r: [r.src, r.dst])
    .collect()
)
#print(edges_array)
```

```python
import toyplot

toyplot.graph(edges_array, width=1000);
```

```python
import toyplot.generate
import toyplot.layout

layout = toyplot.layout.FruchtermanReingold(edges=toyplot.layout.CurvedEdges())
toyplot.graph(edges_array, layout=layout, width=1500);
```

```python
vlstyle = {"fill":"black"}
layout = toyplot.layout.FruchtermanReingold(edges=toyplot.layout.CurvedEdges())
toyplot.graph(edges_array, layout=layout, vcolor="steelblue", vmarker="d", vsize=18, vlstyle=vlstyle, width=1500);
```

```python
colormap = toyplot.color.LinearMap(toyplot.color.Palette(["white", "yellow", "red"]))
vstyle = {"stroke":toyplot.color.near_black}

toyplot.graph(edges_array, layout=layout, vcolor=colormap, vsize=20, vstyle=vstyle, width=1500);
```

```python
estyle = {"stroke-dasharray":"3,3"}
toyplot.graph(
    edges_array,
    layout=layout,
    ecolor="black",
    ewidth=1.2,
    eopacity=0.4,
    estyle=estyle,
    vcolor=colormap,
    vsize=20,
    vstyle=vstyle,
    width=1500,
);
```
